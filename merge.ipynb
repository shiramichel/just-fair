```python
import pandas as pd
import numpy as np
from dfply import *
```

```python
fjc = pd.read_csv('FJC_2018.csv')
ussc = pd.read_csv('mod_ussc_2018.csv')
```

```python
fjc_new = fjc[['ID','SENTMON', 'SENTYR', 'DISTRICT', 'PRISTOT', 'FINETOT', 'PROBTOT', 'FJDcode1', 'FJDcode2', 'FJDcode3', 'FJDcode4', 'FJDcode5', 'FJDcode6', 'FJDcode7', 'FJDcode8', 'FJDcode9', 'FJDcode10']]
ussc_new = ussc[['DISTRICT', 'SENTMON', 'SENTYR', 'SENTTOT_new', 'FINE', 'PROBATN', 'USSCIDN', 'floorSENTTOT', 'roundSENTTOT','USSCcode1', 'USSCcode2', 'USSCcode3', 'USSCcode4', 'USSCcode5']]
ussc_new = ussc_new.rename({'SENTTOT_new': 'PRISTOT', 'FINE': 'FINETOT', 'PROBATN': 'PROBTOT'}, axis=1)
```

```python
fjc_new.replace(-8, 'NA', inplace=True)
fjc_new.replace('-8', 'NA', inplace=True)
fjc_new = fjc_new
```

```python
# first merge on 6 variables
fjc_ussc = pd.merge(ussc_new, fjc_new, on=['SENTMON', 'SENTYR', 'DISTRICT', 'PRISTOT', 'FINETOT', 'PROBTOT'])
fjc_ussc
```

```python
# reformat
fjc_ussc.fillna('NA', inplace=True)
fjc_ussc = fjc_ussc[['ID', 'SENTMON', 'SENTYR', 'DISTRICT', 'PRISTOT', 'FINETOT', 'PROBTOT', 'FJDcode1', 'FJDcode2', 'FJDcode3', 'FJDcode4', 'FJDcode5', 'FJDcode6', 'FJDcode7', 'FJDcode8', 'FJDcode9', 'FJDcode10', 'USSCIDN', 'floorSENTTOT', 'roundSENTTOT', 'USSCcode1', 'USSCcode2', 'USSCcode3', 'USSCcode4', 'USSCcode5']]
fjc_ussc.drop_duplicates(inplace=True)
fjc_ussc
```

```python
# confirm match with codes
matched = []
no_match = []

for i in range(len(fjc_ussc)):
    for j in range(20,25):
        for k in range(7,17):
            if fjc_ussc.iloc[i,k] == fjc_ussc.iloc[i,j] != 'NA':
                matched.append(i)
                break
            else:
                no_match.append(i)
                break
```

```python
# obtain true matches
matched_1 = fjc_ussc.iloc[matched,:]
matched_1.reset_index()
```

```python
# get first unmatched
unmatched = pd.merge(fjc_new, ussc_new, on=['SENTMON', 'SENTYR', 'DISTRICT', 'PRISTOT', 'FINETOT', 'PROBTOT'], how='outer')
unmatched
```

```python
# add unmatched to cases with no code match
unmatched = unmatched.append(fjc_ussc.iloc[no_match,:])
unmatched.drop_duplicates(inplace=True)
unmatched
```

```python
# separate unmatched
unmatched_fjc = (unmatched >>
mask(unmatched.USSCIDN.isna()) >>
drop(unmatched.floorSENTTOT, unmatched.roundSENTTOT, unmatched.USSCIDN, unmatched.USSCcode1, unmatched.USSCcode2, unmatched.USSCcode3, unmatched.USSCcode4, unmatched.USSCcode5)
)

unmatched_ussc = (unmatched >>
mask(unmatched.USSCIDN.notna()) >>
drop(unmatched.ID, unmatched.FJDcode1, unmatched.FJDcode2, unmatched.FJDcode3, unmatched.FJDcode4, unmatched.FJDcode5, unmatched.FJDcode5, unmatched.FJDcode6, unmatched.FJDcode7, unmatched.FJDcode8, unmatched.FJDcode9, unmatched.FJDcode10)
)
```

```python
# check using roundSENTTOT
unmatched_ussc['PRISTOT'] = unmatched_ussc['roundSENTTOT']
```

```python
# change type to merge
unmatched_fjc.PRISTOT = unmatched_fjc.PRISTOT.astype(object)
```

```python
# second merge on 6 variables
fjc_ussc_2 = pd.merge(unmatched_fjc, unmatched_ussc, on=['SENTMON', 'SENTYR', 'DISTRICT', 'PRISTOT', 'FINETOT', 'PROBTOT'])
fjc_ussc_2
```

```python
# confirm second match with codes
matched = []
no_match = []

for i in range(len(fjc_ussc_2)):
    for j in range(20,25):
        for k in range(7,17):
            if fjc_ussc_2.iloc[i,k] == fjc_ussc_2.iloc[i,j] != 'NA':
                matched.append(i)
                break
            else:
                no_match.append(i)
                break
```

```python
# obtain true matches
matched_2 = fjc_ussc_2.iloc[matched,:]
matched_2.reset_index()
```

```python
# merge on unmatched
unmatched_2 = pd.merge(unmatched_fjc, unmatched_ussc, on=['SENTMON', 'SENTYR', 'DISTRICT', 'PRISTOT', 'FINETOT', 'PROBTOT'], how='outer')
unmatched_2
```

```python
# get unmatched
unmatched_2 = unmatched_2.append(fjc_ussc_2.iloc[no_match,:])
```

```python
# separate unmatched

unmatched_fjc_2 = (unmatched_2 >>
mask(unmatched_2.USSCIDN.isna()) >>
drop(unmatched_2.floorSENTTOT, unmatched_2.roundSENTTOT, unmatched_2.USSCIDN, unmatched_2.USSCcode1, unmatched_2.USSCcode2, unmatched_2.USSCcode3, unmatched_2.USSCcode4, unmatched_2.USSCcode5)
)
unmatched_ussc_2 = (unmatched_2 >>
mask(unmatched_2.USSCIDN.notna()) >>
drop(unmatched_2.ID, unmatched_2.FJDcode1, unmatched_2.FJDcode2, unmatched_2.FJDcode3, unmatched_2.FJDcode4, unmatched_2.FJDcode5, unmatched_2.FJDcode5, unmatched_2.FJDcode6, unmatched_2.FJDcode7, unmatched_2.FJDcode8, unmatched_2.FJDcode9, unmatched_2.FJDcode10)
)
```

```python
# check using floorSENTTOT

unmatched_ussc_2['PRISTOT'] = unmatched_ussc_2['floorSENTTOT']
unmatched_ussc_2
```

```python
# merge again
fjc_ussc_3 = pd.merge(unmatched_fjc_2, unmatched_ussc_2, on=['SENTMON', 'SENTYR', 'DISTRICT', 'PRISTOT', 'FINETOT', 'PROBTOT'])
fjc_ussc_3
```

```python
# confirm third match with codes
matched = []
no_match = []

for i in range(len(fjc_ussc_3)):
    for j in range(20,25):
        for k in range(7,17):
            if fjc_ussc_3.iloc[i,k] == fjc_ussc_3.iloc[i,j] != 'NA':
                matched.append(i)
                break
            else:
                no_match.append(i)
                break
```

```python
# get matched
matched_3 = fjc_ussc_3.iloc[matched,:]
matched_3.reset_index()
```

```python
# get unmatched
unmatched_3 = pd.merge(unmatched_fjc_2, unmatched_ussc_2, on=['SENTMON', 'SENTYR', 'DISTRICT', 'PRISTOT', 'FINETOT', 'PROBTOT'], how='outer')
unmatched_3
```

```python
# get unmatched
unmatched_3 = unmatched_3.append(fjc_ussc_3.iloc[no_match,:])
```

```python
# get total matched
matched_2018 = pd.concat([matched_1, matched_2, matched_3], axis=0)
matched_2018.drop_duplicates(inplace=True)
matched_2018
```

```python
# get total unmatched
unmatched_2018 = pd.concat([unmatched, unmatched_2, unmatched_3], axis=0)
unmatched_2018.drop_duplicates(inplace=True)
unmatched_2018
```

```python
# remove duplicates and move to unmatched
unmatched_2018 = pd.concat([matched_2018.loc[matched_2018.duplicated(subset=['USSCIDN'], keep=False)], unmatched_2018], axis=0)
```

```python
# remove duplicates and move to unmatched
unmatched_2018 = pd.concat([matched_2018.loc[matched_2018.duplicated(subset=['ID'], keep=False)], unmatched_2018], axis=0)
```

```python
# drop duplicates
matched_2018 = matched_2018.drop_duplicates(subset=['ID'], keep=False)
matched_2018 = matched_2018.drop_duplicates(subset=['USSCIDN'], keep=False)
```

```python
matched_2018.to_csv('matched_2018.csv', index=False)
```
