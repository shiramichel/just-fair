```python
match_100 = pd.DataFrame()
match_100['USSC'] = ''
match_100['FJC'] = ''
match_100
```

```python
matrix = np.ones((100,100))
matrix
```

```python
for i in range (100):
    for j in range(100):
        if ussc_lim.iloc[i,1] == fjc_new.iloc[j,1] and ussc_lim.iloc[i,2] == fjc_new.iloc[j,2] and ussc_lim.iloc[i,3] == fjc_new.iloc[j,3] and ussc_lim.iloc[i,4] == fjc_new.iloc[j,4] and ussc_lim.iloc[i,5] == fjc_new.iloc[j,5] and ussc_lim.iloc[i,6] == fjc_new.iloc[j,6]:
            match_100.loc[i, 'USSC'] = ussc_lim.iloc[i,0]
            match_100.loc[i, 'FJC'] = fjc_new.iloc[j,0]
            print(i, j)
            matrix[j][i] = 0
        
match_100
```

```python
(unique, counts) = np.unique(matrix, return_counts=True)
frequencies = np.asarray((unique, counts)).T
frequencies
```
