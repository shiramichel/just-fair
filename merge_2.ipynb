```python
q4 = pd.merge(ussc_q4, fjc_q4, on = ['SENTMON', 'SENTYR', 'DISTRICT', 'PRISTOT', 'FINETOT', 'PROBTOT'])
q4 
```

```python
q4.columns
```

```python
matched = []
no_match = []

for i in range(len(q4)):
    for j in range(9,14):
        for k in range(15,25):
            if q4.iloc[i,k] != q4.iloc[i,j]:
                no_match.append(i)
            elif q4.iloc[i,k] == q4.iloc[i,j] != 'NA':
                matched.append(i)
                break # break k
            break # break j
```

```python
matched_1 = q4.iloc[matched,:]
matched_1
```

```python
q4_unmatch = pd.merge(ussc_q4, fjc_q4, on=['SENTMON', 'SENTYR', 'DISTRICT', 'PRISTOT', 'FINETOT', 'PROBTOT'], how='outer')
q4_unmatch
```

```python
unmatched = q4_unmatch.append(q4.iloc[no_match,:])
unmatched.drop_duplicates(inplace=True)
unmatched
```

```python
unmatched_fjc = (unmatched >>
mask(unmatched.USSCIDN.isna()) >>
drop(unmatched.floorSENTTOT, unmatched.roundSENTTOT, unmatched.USSCIDN, unmatched.USSCcode1, unmatched.USSCcode2, unmatched.USSCcode3, unmatched.USSCcode4, unmatched.USSCcode5)
)

unmatched_ussc = (unmatched >>
mask(unmatched.USSCIDN.notna()) >>
drop(unmatched.ID, unmatched.FJDcode1, unmatched.FJDcode2, unmatched.FJDcode3, unmatched.FJDcode4, unmatched.FJDcode5, unmatched.FJDcode5, unmatched.FJDcode6, unmatched.FJDcode7, unmatched.FJDcode8, unmatched.FJDcode9, unmatched.FJDcode10, unmatched.DOCKET)
)
```
