```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
fjc = pd.read_csv('FJC_2018.csv')
ussc = pd.read_csv('ussc_2018_clean.csv')
```

```python
fjc_new = fjc[['ID','SENTMON', 'SENTYR', 'DISTRICT', 'PRISTOT', 'FINETOT', 'PROBTOT', 'FJDcode1', 'FJDcode2', 'FJDcode3', 'FJDcode4', 'FJDcode5', 'FJDcode6', 'FJDcode7', 'FJDcode8', 'FJDcode9', 'FJDcode10']]
ussc_new = ussc[['CIRCDIST', 'SENTMON', 'SENTYR', 'SENTTOT_new', 'FINE', 'PROBATN', 'USSCIDN', 'floorSENTTOT', 'roundSENTTOT','USSCcode1', 'USSCcode2', 'USSCcode3', 'USSCcode4', 'USSCcode5']]
ussc_new = ussc_new.rename({'CIRCDIST': 'DISTRICT', 'SENTTOT_new': 'PRISTOT', 'FINE': 'FINETOT', 'PROBATN': 'PROBTOT'}, axis=1)
```

```python
fjc_new = fjc_new[(fjc_new.PRISTOT != 9996) & (fjc_new.PRISTOT != 9998)]
ussc_new = ussc_new[(ussc_new.PRISTOT != 9996) & (ussc_new.PRISTOT != 9998)]
```

PRISTOT

```python
fig, axs = plt.subplots(1, 1,
                        figsize =(10, 7),
                        tight_layout = True)
axs.hist(fjc_new.PRISTOT)
plt.show()
```
