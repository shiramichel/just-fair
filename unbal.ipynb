```python
import ot
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 15 # number of records in dataset 1 
m = 20 # number of records in dataset 2
k = 8 # number of matching records

a = np.ones(n)/n
b = np.ones(m)/m
M = np.ones((n,m))

for j in range(0,k):
    M[j,j] = 0

epsilon = 0.01 # entropic regularization
alpha = 0.01 #enforces marginals 

epsilon_1 = 0.1
epsilon_2 = 0.001
epsilon_3 = 0.0001

val = []
val_1 = []
val_2 = []
val_3 = []
```

```python
for i in range(100):
    M[0:k, k:m] = np.random.rand(k,m-k)
    M[k:n, 0:k] = np.random.rand(n-k,k)
    M[k:n, k:m] = np.random.rand(n-k,m-k)

    C = ot.unbalanced.sinkhorn_unbalanced(a, b, M, epsilon, alpha)
    C = C/np.max(C)
    val.append(np.max([np.max(C[0:k, 0:k] - np.identity(k)), np.max(C[k:n, 0:k]), np.max(C[0:k, k:m])]))

    C_1 = ot.unbalanced.sinkhorn_unbalanced(a, b, M, epsilon_1, alpha)
    C_1 = C_1/np.max(C_1)
    val_1.append(np.max([np.max(C_1[0:k, 0:k] - np.identity(k)), np.max(C_1[k:n, 0:k]), np.max(C_1[0:k, k:m])]))

    C_2 = ot.unbalanced.sinkhorn_unbalanced(a, b, M, epsilon_2, alpha)
    C_2 = C_2/np.max(C_2)
    val_2.append(np.max([np.max(C_2[0:k, 0:k] - np.identity(k)), np.max(C_2[k:n, 0:k]), np.max(C_2[0:k, k:m])]))

    C_3 = ot.unbalanced.sinkhorn_unbalanced(a, b, M, epsilon_3, alpha)
    C_3 = C_3/np.max(C_3)
    val_3.append(np.max([np.max(C_3[0:k, 0:k] - np.identity(k)), np.max(C_3[k:n, 0:k]), np.max(C_3[0:k, k:m])]))
```

```python
plt.hist(val, bins=10)
plt.ylabel('Frequency')
plt.xlabel('Error Value')
plt.suptitle('15 vs 20 cases; 8 matching')
plt.title('Lambda = 0.01, Rho = 0.01')
```

```python
plt.hist(val_1, bins=10)
plt.ylabel('Frequency')
plt.xlabel('Error Value')
plt.suptitle('15 vs 20 cases; 8 matching')
plt.title('Lambda = 0.1, Rho = 0.01')
```

```python
plt.hist(val_2, bins=10)
plt.ylabel('Frequency')
plt.xlabel('Error Value')
plt.suptitle('15 vs 20 cases; 8 matching')
plt.title('Lambda = 0.001, Rho = 0.01')
```

```python
plt.hist(val_3, bins=10)
plt.ylabel('Frequency')
plt.xlabel('Error Value')
plt.suptitle('15 vs 20 cases; 8 matching')
plt.title('Lambda = 0.0001, Rho = 0.01')
```
