```python
# replace missing values with NaN for modified dataframe

mod.replace(' ', np.NaN, inplace=True)
```

```python
# 2) change SENTTOT to numeric to round and truncate values

mod['SENTTOT'] = pd.to_numeric(mod.SENTTOT, errors='ignore')
```

```python
# change SENTTOT to float

mod['SENTTOT'].astype(float)
```

```python
# create roundSENTTOT round SENTTOT to nearest integer
mod['roundSENTTOT'] = round(mod.SENTTOT)
mod['roundSENTTOT']
```

```python
# create floorSENTTOT to round SENTTOT to hundredths place
mod['floorSENTTOT'] = np.floor(mod.SENTTOT)
```

```python
# 3) reconcile the two variables 

def rec(row):
    if row['TOTPRISN'] == 9996:
        return 9996
    elif row['TOTPRISN'] == 9998:
        return 9998
    return row['SENTTOT']
```

```python
mod['SENTTOT_new'] = mod.apply(rec, axis=1)
```

```python
# 4) set NaN to 0 to indicate zero-length sentences
mod.SENTTOT.fillna(0, inplace=True)
mod.SENTTOT_new.fillna(0, inplace=True)
```

```python
# create SENTTOTFLAg to flag records later

mod['SENTTOTFLAG'] = ''
```

```python
# create function to identify index of first occuring letter and truncate before that index

def trunc(code):
    res = None
    temp = re.search(r'[a-z]', code, re.I)
    if temp is not None:
        res = temp.start()
        return code[:res]
    else:
        return code
```
